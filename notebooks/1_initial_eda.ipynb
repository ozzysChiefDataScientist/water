```python
##############
# Perform cleanup of water system ID columns
###############
```

```python
# This column has whitespace that needs to be removed
print("Original values in prod PWSID: {}".format(prod['PWSID'].values[0:5]))
prod['PWSID'] = prod['PWSID'].apply(lambda x: x.rstrip().lstrip())

print("Original values in prod PWSID: {}".format(deliv['PWSID'].values[0:5]))
deliv['PWSID'] = deliv['PWSID'].apply(lambda x: x.rstrip().lstrip())

print("Original values in systems Water System No: {}".format(systems['Water System No'].values[0:5]))
systems['Water System No'] = systems['Water System No'].apply(lambda x: str(x).rstrip().lstrip())
```

```python
###########
# Search for non-numerical units in numeric columns
###############
```

```python
numeric_prod_col = prod.columns.values[11:-2]
numeric_deliv_col = deliv.columns.values[11:-2]

print("Names of numeric columns in prod: {}".format(numeric_prod_col))
print("***")
print("Names of numeric columns in deliv: {}".format(numeric_deliv_col))
```

```python
# how are numeric values stored?
# strings with extra whitespace; commas; decimal points
print("Sample numbers in prod numeric columns: {}".format(prod['WATER PRODUCED FROM GROUNDWATER'].values))

print("Sample numbers in deliv numeric columns: {}".format(deliv['WATER DELIVERIES Total.Delivered Residential IN REVISED UNITS (Total Does not include Landscape Irrigation, Agricultural or to other PWS)'].values))
```

```python
# convert numeric columns from string to numeric

for c in numeric_prod_col:
    print(c)
    prod[c+"_numeric"] = prod[c].apply(lambda x: pd.to_numeric(str(x).replace(",",""),errors='coerce'))

for c in numeric_deliv_col:
    print(c)
    deliv[c+"_numeric"] = deliv[c].apply(lambda x: pd.to_numeric(str(x).replace(",",""),errors='coerce'))
```

```python
############
# Prepare to convert units
#############
```

```python
prod_meta_col = list(prod.columns.values[0:10]) + ['CALCULATED GPCD (Total Potable Produced in gallons per capita day)','Population Of Service Area']
print(prod_meta_col)

deliv_meta_col = list(deliv.columns.values[0:10]) + ['CALCULATED GPCD (Total delivery to residential in gallons per capita day)','Population Of Service Area']
print(deliv_meta_col)
```

```python
# Identify columns with water-related numbers (not population counts)
prod_water_numeric_col = [x+"_numeric" for x in numeric_prod_col if x not in prod_meta_col]
print(prod_water_numeric_col)
print("***")
deliv_water_numeric_col = [x+"_numeric" for x in numeric_deliv_col if x not in deliv_meta_col]
print(deliv_water_numeric_col)
```

```python
# Melt so we have one row per different water numeric column
prod_melt = pd.melt(prod[prod_meta_col+list(prod_water_numeric_col)], id_vars=list(prod_meta_col), value_vars=list(prod_water_numeric_col))

deliv_melt = pd.melt(deliv[deliv_meta_col+list(deliv_water_numeric_col)], id_vars=list(deliv_meta_col), value_vars=list(deliv_water_numeric_col))
```

```python
############
# Convert units
# Explore units
#AF=Acre Feet, 
#CCF=Hundred Cubic Feet, 
#MG=Million Gallons
#G=Gallons, 
#HG=Hundred Gallons
#DG=Tenth Gallons, 
#CF=Tenth Cubic Feet
#-=NO UNITS REPORTEDOR OIMA REMOVED THE UNITS
#############
```

```python
prod_melt['WATER PRODUCED Water.Units REVIEWED BY OFFICE OF INFORMATION MANAGEMENT AND ANALYSIS'].value_counts(dropna=False)
```

```python
# drop rows where the units are not recorded
prod_melt = prod_melt[prod_melt['WATER PRODUCED Water.Units REVIEWED BY OFFICE OF INFORMATION MANAGEMENT AND ANALYSIS']!="-"]
print(prod_melt.shape)

deliv_melt = deliv_melt[deliv_melt['Delivered.Water.Units.Revised BY OFFICE OF INFORMATION MANAGEMENT AND ANALYSIS']!="-"]
print(deliv_melt.shape)
```

```python
# key = original unit
# value = multiplication factor to obtain gallons
conversionDict = {"AF": 325851.43,
                  "CF": 7.48051948,
                "CCF": 748.051948,
                  "DG": 10,
                  "G":1,
                  "HG": 100.,
                  "MG": 1000000.,
                 "TG": .1}
```

```python
# convert units to gallons based on conversionDict
prod_melt['value_gallons'] = prod_melt[['value',
      'WATER PRODUCED Water.Units REVIEWED BY OFFICE OF INFORMATION MANAGEMENT AND ANALYSIS']].apply(lambda x: conversionDict[x[1]]*x[0],axis=1)

# convert units to gallons based on conversionDict
deliv_melt['value_gallons'] = deliv_melt[['value',
      'Delivered.Water.Units.Revised BY OFFICE OF INFORMATION MANAGEMENT AND ANALYSIS']].apply(lambda x: conversionDict[x[1]]*x[0],axis=1)
```

```python
# Pivot, but do not include units columns since units are now standardized to gallons
deliv_pivot = deliv_melt.pivot_table(index=[x for x in list(deliv_meta_col) if "units" not in x.lower()], columns='variable', values='value_gallons')
deliv_pivot = deliv_pivot.reset_index()

prod_pivot = prod_melt.pivot_table(index=[x for x in list(prod_meta_col) if "units" not in x.lower()], columns='variable', values='value_gallons')
prod_pivot = prod_pivot.reset_index()
```

```python
# count the number of unique water systems per PWSID/Date
ctWaterSystems = prod_pivot.groupby(['PWSID','Date']).agg({"Water.System.Classification": lambda x: x.nunique()}).reset_index()
```

```python
# Which PWSID/Dates have multiple water system classifications?
multipleClassifications = ctWaterSystems[ctWaterSystems['Water.System.Classification']>1]
multipleClassifications['key'] = multipleClassifications[['Date','PWSID']].apply(lambda x: "{}___{}".format(x[0],x[1]),axis=1)
multipleClassifications
```

```python
# Change Water System Classification to NA when a PWSID/Date has multiple water system classifications
prod_pivot['key'] = prod_pivot[['Date','PWSID']].apply(lambda x: "{}___{}".format(x[0],x[1]),axis=1)
prod_pivot.loc[prod_pivot['key'].isin(multipleClassifications['key']),'Water.System.Classification'] = np.nan
```

```python
# Drops duplicate rows that previously existed due to multiple water system classifications being associated with
# one PWSID/Date
print(prod_pivot.shape)
prod_pivot = prod_pivot.drop_duplicates()
print(prod_pivot.shape)
```

```python
# the PWSID/Date is now the unique key
rowCounts = prod_pivot.groupby(['PWSID','Date']).count().reset_index()
rowCounts[rowCounts['Water.System.Name']>1][0:4]
```
